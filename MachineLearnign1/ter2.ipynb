```python
import pandas as pd
from sklearn.tree import DecisionTreeRegressor
```

```python
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split
```

```python
iowa_path ='train.csv'
home_data=pd.read_csv(iowa_path)
```

```python
home_data.describe()
```

```python
y= home_data.SalePrice
```

```python
feature_colums =['LotArea', 'YearBuilt', '1stFlrSF', '2ndFlrSF', 'FullBath', 'BedroomAbvGr', 'TotRmsAbvGrd']
```

```python
x= home_data[feature_colums]
```

```python
iowa_model =DecisionTreeRegressor()
iowa_model.fit(x,y)
```

```python
print('Primera predicciÃ²n:', iowa_model.predict(x.head()))
print('Valor real',y.head().tolist())
```

```python
train_x,value_x,train_y,value_y = train_test_split(x,y,random_state=0)
```

```python
iowa_model= DecisionTreeRegressor()
iowa_model.fit(train_x,train_y)
```

```python
val_predictions =iowa_model.predict(value_x)
```

```python
print('Prediccion: ',val_predictions)
print('valor real',value_y.tolist())
```

```python
val_mae=mean_absolute_error(value_y,val_predictions)
```
